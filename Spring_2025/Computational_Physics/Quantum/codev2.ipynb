```python
#define spacing and x,t arrays
dx=1
dt=1
xi=0
xF=10
t0=0
tF=10
x0=np.linspace(0,5,100)
t=np.linspace(t0,tF,11)

#define initial phi
phi=np.exp(-2*(x0-2.5)**2)
#define empty matrix where we hold x
x_mat=[]

#define A
gamma=.1
B=np.zeros(len(x0))
B[0]=1
B[-1]=1
B[1:-1]=(1+2*gamma)
A=np.diag(B)

#solve for inital x
sol=np.linalg.solve(A,phi)
#print(sol.shape)
x_mat.append(sol)
#x_mat
sol2=np.linalg.solve(A,sol)
x_mat.append(sol2)
x_mat[-1]
#print(np.shape(x_mat))
```

```python
#define empty matrix where we hold x
x_mat=[]
sol=np.linalg.solve(A,phi)
x_mat.append(sol)

for i in range(10):
    sol=np.linalg.solve(A,x_mat[i])
    x_mat.append(sol)
x_mat
print(len(x_mat))
```

```python
plt.figure()
for i in range(len(x_mat)):
    plt.plot(x0,x_mat[i], label="Time {}".format(i))
plt.xlabel("Position (x)")
plt.ylabel("Phi")
plt.legend()
plt.show()
```
