```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler()->tuple:
      """
      Euler Implementation:
      Parameters
      ----------
      f : Function
            Quadradic drag function
      t0 : Float
            Initial time
      tN : Float
            Final time
      N : Integer
            Number of steps
      """
      t = np.linspace(t0, tN, N)
      v    = np.zeros([N])
      y    = np.zeros([N])
      h = (tN-t0)/N
      for i in range(1,N):
            v[i] = v[i-1] + h*f(v[i-1])
            y[i] = y[i-1] + h*v[i]
      return t, y, v
```
